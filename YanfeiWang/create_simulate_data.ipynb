## Generated simulation data for the data product version 2

```python
# add missing values for protocol and conn state for later one-hot encode
import pandas as pd
import csv
IOT23_COLUMNS = ['ts',
              'uid',
              'id.orig_h',
              'id.orig_p',
              'id.resp_h',
              'id.resp_p',
              'proto',
              'service',
              'duration',
              'orig_bytes',
              'resp_bytes',
              'conn_state',
              'local_orig',
              'local_resp',
              'missed_bytes',
              'history',
              'orig_pkts',
              'orig_ip_bytes',
              'resp_pkts',
              'resp_ip_bytes',
              'label']
filepath = "conn.log.labeled.raw"
df_pre = pd.read_table(filepath_or_buffer=filepath, skiprows=8, nrows=10)
df_pre.columns=IOT23_COLUMNS
dummy_row=df_pre.iloc[1].to_dict()
# add missing types for the dataset
dummy_uid='WILLBEDELETED000'
unique_proto = ['icmp', 'udp', 'tcp']
unique_conn_state = ['OTH','REJ','RSTO','RSTOS0','RSTR','RSTRH','S0','S1','S2','S3','SF','SH','SHR']
dummy_rows=[]
for i in range(len(unique_conn_state)):
    row = dummy_row.copy()
    postfix = '0'+str(i) if i < 10 else str(i)
    row['uid'] = 'WILLBEDELETED000%s'%postfix
    if i < 3:
        row['proto'] = unique_proto[i]
    row['conn_state'] = unique_conn_state[i]
    dummy_rows.append(row)
with open('dummy.csv', 'w') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=df_pre.columns,delimiter='\x09')
    writer.writerows(dummy_rows)

df_dummy = pd.read_table(filepath_or_buffer='dummy.csv', header=None)
df_dummy.columns=IOT23_COLUMNS
df_dummy.head()
```

```python
# load raw data
df_c = pd.read_table(filepath_or_buffer=filepath, header=None,skiprows=8)
df_c.columns=IOT23_COLUMNS
df_c.drop(df_c.tail(1).index,inplace=True)
df_c = pd.concat([df_dummy, df_c], ignore_index=True)
df_c = df_c.set_index('uid')
df_c.head()
```

```python
# drop unused features
df = df_c.drop(columns=['ts','id.orig_h','id.orig_p','id.resp_h','id.resp_p',
                               'local_orig','local_resp','history','service'])
```

```python
# filter labels
def process_label(s):
    _, label, detail = s.split()
    if label == 'Benign':
        return label
    else:
        return detail
df['label'] = df['label'].apply(process_label)
df = df[(df['label'] == 'PartOfAHorizontalPortScan') | (df['label'] == 'Okiru') 
            | (df['label'] == 'Benign') | (df['label'] == 'DDoS')]
df['label'].value_counts()
```

```python
# one-hot encode
df = pd.get_dummies(df, columns=['proto'])
df = pd.get_dummies(df, columns=['conn_state'])
```

```python
# handle na values
df['duration'] = df['duration'].str.replace('-','0')
df['orig_bytes'] = df['orig_bytes'].str.replace('-','0')
df['resp_bytes'] = df['resp_bytes'].str.replace('-','0')
df.fillna(-1,inplace=True)
# randomly select 3000 records for production
df = df.sample(n = 3000)
```

```python
# generate simulation data
series = df.apply(
    lambda x: ','.join(x.dropna().astype(str)),
    axis=1)
df_ready = pd.concat([series, df_c], axis=1, join="inner")
df_ready.columns
df_ready['data']=df_ready[0]
df_ready=df_ready.drop(columns=[0])
df_ready.reset_index()
df_ready.to_csv('simulate_data.csv')
```
